```python
import pandas as pd
data=pd.read_csv("/content/indian-state-level-election.csv")
data.isnull().sum()
data
```

**Lets drop the null values rows**

---



```python
data.dropna(subset=['ac_name', 'partyname', 'partyabbre', 'cand_sex', 'totvotpoll'], inplace=True)
data.reset_index(drop=True, inplace=True)
data
```

**If i want to know who rules Andra**

---



```python
AndhraCollection = data[data['st_name'] == 'Andhra Pradesh']
AndhraCollection
```

```python
sortedAndra=AndhraCollection.sort_values(by='totvotpoll',ascending=False)
LeadInAndhra=sortedAndra.iloc[0]
LeadInAndhra['partyname']
```

**Describing**

```python
data['totvotpoll'].describe()
```

  **To see which state secures highest vote**

---



```python
max_votes = data['totvotpoll'].max()
state_with_max_votes = data[data['totvotpoll'] == max_votes]['st_name'].values[0]

print(f"Highest vote count: {max_votes} in {state_with_max_votes}")
```

**Removing duplicates**

---



```python
data = data.loc[:, ~data.columns.duplicated()]
data
```

**SEEING OUTLIERS**

---



```python
import seaborn as sns
import matplotlib.pyplot as plt


sns.boxplot(data=data, x='totvotpoll')
plt.title('Box Plot of totvotpoll')
plt.show()


sns.boxplot(data=data, x='electors')
plt.title('Box Plot of electors')
plt.show()
data
```

**outliers removal**

---



```python
import pandas as pd
from scipy import stats
import numpy as np
z_scores = np.abs(stats.zscore(data['totvotpoll']))
threshold = 2

condition = z_scores < threshold

data = data[condition]

print(data)
filtered_data = data[condition].reset_index(drop=True)
print(filtered_data)

sns.boxplot(data=filtered_data, x='totvotpoll')
plt.title('Box Plot of totvotpoll (after outlier removal)')
plt.show()
```

***Graphs***

---



```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
plt.figure(figsize=(10, 6))
sns.histplot(data['totvotpoll'], kde=True)
plt.title('Distribution of Total Votes Polled')
plt.xlabel('Total Votes Polled')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(data=data, x='totvotpoll', y='electors')
plt.title('Scatter Plot of Total Votes Polled vs. Electors')
plt.xlabel('Total Votes Polled')
plt.ylabel('Electors')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
correlation_matrix = data.corr()
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()
```
